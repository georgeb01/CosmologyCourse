# Proper distances in single-component universes

In Lecture 7 we consider a variety of single-component universes (i.e., ones that *only* contain either matter, photons or dark energy). While these clearly don't represent the real Universe, they are important for two reasons. Firstly, they're important from a historical perspective since these were the first types of universe to be solved analytically. Secondly, from a physical perspective: during most of the history of the Universe, it has been dominated by a single component. The dominant component has changed over time: first, it was dominated by photons, then matter and now, dark energy. As such, single component universes provide a fairly good approximation for most of the history of the real Universe.

In the following, I provide the code that I used to generate the plots in Lecture 7 that describe the relationships between proper distance and redshift in different single-component universes. As with all code associated with the course, please feel free to clone and play around with them.

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.logspace(-2,2.5)
c_H0 = 1.

#Proper distance at the time of observation:
#-------------------------------------------

#For a matter-only model universe, omega = 0
omega = 0.
expo = -1 * (1+3*omega)/2
zterm = (1+z)**expo
dp_tob = c_H0 * (2/(1+3*omega)) * (1-zterm)
plt.plot(z, dp_tob)

#For a photon-only model universe, omega = 1/3
omega = 1./3.
expo = -1 * (1+3*omega)/2
zterm = (1+z)**expo
dp_tob = c_H0 * (2/(1+3*omega)) * (1-zterm)
plt.plot(z, dp_tob, '--')

#For a dark energy-only model universe, omega = 1/3
dp_tob = c_H0 * z
plt.plot(z, dp_tob, '-.')

plt.xscale('log')
plt.yscale('log')
plt.axis([0.01,300,0.01,30])
plt.xlabel('z')
plt.ylabel('Proper distance (units: c/H0)')
plt.title('Proper distance at time of observation')

plt.show()
```

```python
#Proper distance at the time of emission:
#-------------------------------------------

#For a matter-only model universe, omega = 0
omega = 0.
expo = -1 * (1+3*omega)/2
zterm = (1+z)**expo
dp_tob = c_H0 * (2/(1+3*omega)) * (1-zterm)
plt.plot(z, dp_tob/(1+z))

#For a photon-only model universe, omega = 1/3
omega = 1./3.
expo = -1 * (1+3*omega)/2
zterm = (1+z)**expo
dp_tob = c_H0 * (2/(1+3*omega)) * (1-zterm)
plt.plot(z, dp_tob/(1+z), '--')

#For a dark energy-only model universe, omega = 1/3
dp_tob = c_H0 * z
plt.plot(z, dp_tob/(1+z), '-.')

plt.xscale('log')
plt.yscale('log')
plt.axis([0.01,300,0.01,30])
plt.xlabel('z')
plt.ylabel('Proper distance (units: c/H0)')
plt.title('Proper distance at time of emission')

plt.show()
```
