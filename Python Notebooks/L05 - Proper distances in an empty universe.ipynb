# Proper distances in an empty universe

During the lecture, we derived the relationship between redshift and proper distances for an empty universe. In the lecture, the plots showing these relationships were presented. Below is the code used to generate these plots:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

z = np.logspace(-2,2.5)

c_H0 = 1.
dp_tobs = c_H0 * np.log(1.+z)

#Proper distance at time of observation
plt.plot(z, dp_tobs)
plt.xscale('log')
plt.yscale('log')
plt.axis([0.01,300,0.01,30])
plt.xlabel('z')
plt.ylabel('Proper distance (units: c/H0)')
plt.title('Proper distance at time of observation')
plt.show()
```

```python
#Proper distance at time of emission:
plt.plot(z, dp_tobs/(1.+z))
plt.xscale('log')
plt.yscale('log')
plt.axis([0.01,300,0.01,30])
plt.xlabel('z')
plt.ylabel('Proper distance (units: c/H0)')
plt.title('Proper distance at time of emission')
plt.show()
```
