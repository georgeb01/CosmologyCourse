# Lecture 9: The CMB Part 1
The following code is what I used to generate the plots shown in slides for Lecture 9. Nothing particularly special or complicated.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling.blackbody import blackbody_nu
from astropy import units as u

en = np.logspace(-2,1.6,1028) * u.eV
temp = 20000. * u.K
bb = blackbody_nu(en, temp)
nphot = bb/en

fig=plt.figure(figsize=(9, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size':15})
plt.plot(en, nphot/np.max(nphot))
plt.xlabel("Energy/eV")
plt.ylabel("Relative number of photons")

#plt.xscale('log')
#plt.yscale('log')

plt.axis([0,38,0,1.1])
```

```python
import numpy as np
import matplotlib.pyplot as plt


T = np.logspace(2.8,5,1028)
me = 9.1e-31
c2 = 9e16
Q = 13.6*1.6e-19
k = 1.4e-23
eta = 6.1e-10

p1 = 3.84*eta
p2 = (k*T)/(me*c2)
p3 = np.exp(Q/(k*T))
x_x2 = p1*(p2**(3./2.))*p3

fig=plt.figure(figsize=(9, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size':15, 'text.usetex':True})

plt.plot(T, x_x2)
plt.xscale('log')
plt.yscale('log')

plt.xlabel("Temperature/K")
plt.ylabel("(1-X)/X^2")

plt.axis([0.8e3,1.2e4,0.1,100])
```
